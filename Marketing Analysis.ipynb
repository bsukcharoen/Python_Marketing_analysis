# Cleaning, joining data

 **Fix Google Analytics Page data** 

```python
ga_page_data
```

```python
#split the url by /
split_domain = ga_page_data.Page.str.rpartition("/")
```

```python
split_domain
```

```python
split_domain[2].replace("","miratrix.co.uk",inplace =True)
split_domain
```

```python
#build the URL and replace current URL's with new URL format
ga_page_data.Page = "http://" + split_domain[2]+split_domain[0]+"/"
ga_page_data
```

- As you can see the **Avg.time on page** which is the time series.So it shouldn't be the object type , it should be time datatype.

- **Bounce Rate** should be float or int
- **% exit** should be float
- **Page value** should be float

```python
ga_page_data.info()
```

**Check the type of data to be able to use for further analysis**
 - Avg. Time on Page
 - Bounce Rate
 - % Exit
 - Page Value

```python
#Replace 0 with time format 00:00:00
ga_page_data['Avg. Time on Page'] = ga_page_data['Avg. Time on Page'].replace(0,'00:00:00')

#convert column from string to time delta
ga_page_data['Avg. Time on Page'] = pd.to_timedelta(ga_page_data['Avg. Time on Page'])
```

```python
#Replace % with nothing
ga_page_data['Bounce Rate'] = ga_page_data['Bounce Rate'].str.replace('%',"")

#convert the column into a float
ga_page_data['Bounce Rate'] = ga_page_data['Bounce Rate'].astype('float')
```

```python
#Replace % with nothing
ga_page_data['% Exit'] = ga_page_data['% Exit'].str.replace('%',"")

#convert column to float type
ga_page_data['% Exit'] =ga_page_data['% Exit'].astype('float')
```

```python
# Replace % with nothing
ga_page_data['Page Value'] = ga_page_data['Page Value'].str.replace('£',"")

#convert type to float
ga_page_data['Page Value'] = ga_page_data['Page Value'].astype('float')
```

```python
#Set the index for df to Page
ga_page_data.set_index('Page',inplace = True)
```

```python
#Check the data after changed the type of data
ga_page_data.info()
```

**Apply Groupby function**

```python
# groupbt Index and get the mean of the 'Avg Time on Page'
avg_time = ga_page_data.groupby(ga_page_data.index)['Avg. Time on Page'].mean(numeric_only = False) #false because we're dealing with datetime not a number
```

```python
#convert avg_time to df and store as ga_time
ga_time = pd.DataFrame(avg_time)
```

```python
#groupby bounce , exit, pagevies , unique pageviews using the appropriate math
#store in a dataframe , save as ga_bou_ex_type

ga_bou_ex_type = pd.DataFrame(ga_page_data.groupby(ga_page_data.index).agg({
    'Bounce Rate' : ['mean'],
    '% Exit'      : ['mean'],
    'Pageviews'   : sum,
    'Unique Pageviews' : sum
}))
```

```python
#check time - Groupby
ga_time.head()
```

```python
ga_bou_ex_type.info()
```

**Rebuilding the Google Analytics Dataframe**

```python
new_ga = pd.merge(ga_time,         #left dataset
                 ga_bou_ex_type,   #right dataset
                 how = 'inner',    #joining using the data in the column
                 left_on = 'Page', #the left dataset is neing joined on the column page
                 right_on = 'Page').sort_values(by='Avg. Time on Page',ascending = False) #the right dataset is neing joined on the column Page)
```

```python
new_ga.head()
```

```python
#rename(Bounce Rate, mean),(% Exit,mean),(Pageviews,sum),(Unique Pagevies, sum)
new_ga = new_ga.rename(columns = {('Bounce Rate','mean'): 'Bounce Rate',
                                 ('% Exit','mean'):'% Exit',
                                 ('Pageviews','sum'):'Pageviews',
                                 ('Unique Pageviews','sum'):'Unique Pageviews'})
                       
```

## matplotlib

- Plotting two y-axes
- create x , y label
- rotating xtick label
- create the legend()
- adding title
- adding annotation 

```python
fig = plt.figure(figsize = (15,5))                 #create matplotlib figure

ax = fig.add_subplot(111)                          #create matplotlib axes
ax2 = ax.twinx()                                   #create another axes that shares the same x-axis as ax

width = 0.2                                        #width of the bar


ga_gsc_data['Clicks'][:10].plot(kind = 'bar',
                                color = 'red',      #color of bar
                                ax=ax,              #what plt axes to use
                                width = width,      #position of bar relative to the others
                                position = 1,
                                fontsize = 10       #set font size
                               )   


ga_gsc_data['Pageviews'][:10].plot(kind = 'bar',    #type of plot
                                 color = 'lightgreen',
                                 ax=ax2,
                                 width = width,
                                 position = 0)     #position of the bar relative of the others



#create ax customisations

#create annotation
ax.annotate('Lots of Pageviews but few clicks',
           xy = (8,40),
           xytext = (2,120),
           arrowprops = dict(facecolor = 'cyan',    #color
                            shrink = 0.05,         #length of arrow
                            lw = 1,                 #line width
                            ec= 'black',          #boarder colour
                            zorder =1))             #layering order of annotation


ax.set_xlabel('Page')                               #name the x axis
ax.set_ylabel('Clicks')                             #name the primary y axis

#rotating xtick label
ax.set_xticklabels(ax.get_xticklabels(),            #str with the label names
                  rotation = -45)                   #rotate the xticks labels by degree



#create ax2 customize
ax2.annotate('Lots of Pageviews and clicks',
            xy=(0,1398),
            xytext = (1,1398),
            arrowprops = dict(facecolor = 'cyan',    #color
                            shrink = 0.05,         #length of arrow
                            lw = 1,                 #line width
                            ec= 'black',          #boarder colour
                            zorder =1))             #layering order of annotation


ax2.set_ylabel('Pageviews')                         #name the secondary y axis



#global plot setting
fig.legend()                                        #set the legend

#create the title of the plt
plt.title('Bar Chart: Clicks vs Pageview')      



plt.show()
```

**Plot weekly PPC CPC comparison**
